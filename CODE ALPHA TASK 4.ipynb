## Task 4: Sentiment Analysis

### Project Title:
Amazon Product Reviews Sentiment Analysis

### Objective:
The objective of this project is to analyze Amazon customer reviews and classify them into Positive, Negative, and Neutral sentiments using Natural Language Processing (NLP).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
!pip install textblob
from textblob import TextBlob
df = pd.read_csv('/content/amazon.csv')
df = df[['product_name','review_title','review_content','rating']]

df.head()
```

```python
# CHECK MISSING VALUES
df.isnull().sum()
```

```python
print(df.shape)
```

```python
# CREATE SENTIMENT FUNCTION
def get_sentiment(text):

    polarity = TextBlob(str(text)).sentiment.polarity

    if polarity > 0:
        return "Positive"

    elif polarity < 0:
        return "Negative"

    else:
        return "Neutral"
# SENTIMENT FUNCTION
df['Sentiment'] = df['review_content'].apply(get_sentiment)

df.head()
```

```python
# COUNT SENTIMENTS
df['Sentiment'].value_counts()
```

```python
# SENTIMENT BAR CHART
plt.figure(figsize=(6,4))

sns.countplot(x='Sentiment', data=df)

plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")

plt.show()
```

### Interpretation

The chart shows the distribution of Positive, Negative, and Neutral reviews. It helps understand overall customer satisfaction with Amazon products.

```python
# SENTIMENT PIE CHART
sentiment_counts = df['Sentiment'].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    sentiment_counts,
    labels=sentiment_counts.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Sentiment Distribution")

plt.show()
```

### Interpretation

The pie chart represents the percentage of positive, negative, and neutral customer reviews. It provides a quick overview of customer opinion.

```python
# TOP 5 POSITIVE REVIEWS
df[df['Sentiment'] == 'Positive'][['product_name', 'review_title']].head()
```

```python
# TOP 5 NEGATIVE REVIEWS
df[df['Sentiment'] == 'Negative'][['product_name', 'review_title']].head()
```
