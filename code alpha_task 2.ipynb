```python
# IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
df = pd.read_csv("/content/amazon.csv")
df.head()
```

## Questions Before Analysis

1. Which product categories have the highest number of products?
2. What is the average product rating?
3. Which products have the highest discount percentage?
4. Does a higher discount result in better ratings?
5. Which products have the highest number of customer ratings?
6. Are there any missing values or duplicate records?
7. What trends can be identified from the dataset?

```python
# EXPLORE DATA STRUCTURE
print(df.shape)

df.columns

df.info()

df.describe(include='all')
```

```python
# CHECK MISSING VALUES
df.isnull().sum()
```

```python
# Remove missing values
df.dropna(subset=['rating', 'rating_count'], inplace=True)

# Check again
print(df.isnull().sum())

# Check shape
print(df.shape)
```

```python
# CHECK DUPLICATE VALUES
df.duplicated().sum()
```

DATA CLEANING

```python
print(df.shape)

df.columns

df.info()

df.describe(include='all')
```

```python
# VARIFIED CLEANED DATA
print(df.shape)

df.columns

df.info()

df.describe(include='all')
```

```python
# RATING DISTRIBUTION
plt.figure(figsize=(8,5))

plt.hist(df['rating'], bins=10)

plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```

```python
# DISCOUNT DISTRIBUTION
plt.figure(figsize=(8,5))

plt.hist(df['discount_percentage'], bins=20)

plt.title("Discount Percentage Distribution")
plt.xlabel("Discount Percentage")
plt.ylabel("Count")

plt.show()
```

```python
# TOP 10 CATEGORIES
plt.figure(figsize=(12,6))

df['category'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")

plt.xticks(rotation=90)

plt.show()
```

```python
# ACTUAL PRICE AND DISCOUNTED PRICE
# Data type conversion (re-applying the cleaning logic)
df['discounted_price'] = df['discounted_price'].apply(lambda x: str(x).replace('₹', '').replace(',', '')).astype(float)
df['actual_price'] = df['actual_price'].apply(lambda x: str(x).replace('₹', '').replace(',', '')).astype(float)
df['discount_percentage'] = df['discount_percentage'].apply(lambda x: str(x).replace('%', '')).astype(float)
df['rating'] = df['rating'].apply(lambda x: str(x))
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = df['rating_count'].apply(lambda x: str(x).replace(',', ''))
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

filtered = df[df['actual_price'] < 20000]

plt.figure(figsize=(8,6))

plt.scatter(
    filtered['actual_price'],
    filtered['discounted_price'],
    alpha=0.5,
    s=20
)

plt.title("Actual Price vs Discounted Price (Filtered)")
plt.xlabel("Actual Price")
plt.ylabel("Discounted Price")

plt.show()
```

CORRELATION (HYPOTHESIS TESTING)

```python
correlation = df[['actual_price',
                  'discounted_price',
                  'discount_percentage',
                  'rating',
                  'rating_count']].corr()

print(correlation)
```

```python
# CORRELATION HEATMAP
correlation = df[['actual_price',
                  'discounted_price',
                  'discount_percentage',
                  'rating',
                  'rating_count']].corr()

print(correlation)
```

```python
# DETECT POTENTIAL DATA ISSUES
print("Missing Values")
print(df.isnull().sum())

print("\nDuplicate Rows")
print(df.duplicated().sum())

print("\nData Types")
print(df.dtypes)
```
